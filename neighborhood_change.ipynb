```python
all_counties_gdf = gpd.read_file("./data/shapes/tl_2020_us_county")

va_tracts_gdf = gpd.read_file("./data/shapes/tl_2020_51_tract")
md_tracts_gdf = gpd.read_file("./data/shapes/tl_2020_24_tract")
dc_tracts_gdf = gpd.read_file("./data/shapes/tl_2020_11_tract")
```

```python
# get just the county shapes that border DC

va_county_names = ["Loudoun", "Arlington", "Falls Church", "Alexandria", "Fairfax"]
md_county_names = ["Montgomery", "Prince George's"]

va_county_gdf = all_counties_gdf[(all_counties_gdf["STATEFP"] == "51") & (all_counties_gdf["NAME"].isin(va_county_names))]
md_county_gdf = all_counties_gdf[(all_counties_gdf["STATEFP"] == "24") & (all_counties_gdf["NAME"].isin(md_county_names))]
```

```python
# get census tracts from the target counties bordering DC

target_va_tracts_gdf = gpd.sjoin(va_tracts_gdf, va_county_gdf, how="inner", predicate="within")
target_md_tracts_gdf = gpd.sjoin(md_tracts_gdf, md_county_gdf, how="inner", predicate="within")
```
